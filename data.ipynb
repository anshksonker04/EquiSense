```python
import yfinance as yf
import pandas as pd

# Download 2 years of Apple stock data
print("Downloading Apple Stock Data...")
apple_data = yf.download("AAPL", start="2024-01-01", end="2026-01-01")

# Clean it up
apple_data.dropna(inplace=True)

# Show the results on the screen
display(apple_data.head())
```

```python
# Testing the NLP Brain
import torch
from transformers import pipeline

print("Downloading the FinBERT AI Brain... (This happens only once)")

# Load the financial sentiment analyzer
sentiment_analyzer = pipeline("sentiment-analysis", model="ProsusAI/finbert")

# Let's feed it some fake, realistic financial headlines
headlines = [
    "Apple reports record-breaking revenue for the third quarter, beating all analyst expectations.",
    "Supply chain issues and factory strikes cause massive delays for Apple's new product launch.",
    "Apple CEO Tim Cook to speak at upcoming tech conference in California."
]

print("\n--- AI Reading the News ---")
for headline in headlines:
    result = sentiment_analyzer(headline)[0]
    
    label = result['label']
    confidence = result['score'] * 100
    
    print(f"\nHeadline: '{headline}'")
    print(f"AI Prediction: {label.upper()} (Confidence: {confidence:.2f}%)")
```

```python
# Connecting the brain to the Live Internet
import yfinance as yf
import pandas as pd

print("Fetching live news for Apple from Yahoo Finance...\n")
# 1. Connect to the stock ticker
aapl = yf.Ticker("AAPL")
live_news = aapl.news

# 2. Debugging: Let's peek at the actual data structure Yahoo gave us
if len(live_news) > 0:
    print(f"The actual keys Yahoo sent for the first article: {list(live_news[0].keys())}")
else:
    print("Warning: Yahoo returned an empty list. No news found.")

# 3. Defensive Extraction
real_headlines = []
for article in live_news:
    # Safely check for the title in different possible locations
    if 'title' in article:
        real_headlines.append(article['title'])
    # Sometimes Yahoo nests it inside a 'content' dictionary
    elif 'content' in article and 'title' in article['content']:
        real_headlines.append(article['content']['title'])

print("\n--- Live Market Sentiment Analysis ---")
# 4. Feed the safely extracted headlines into your FinBERT brain
for headline in real_headlines[:5]:
    # We use the 'sentiment_analyzer' from the previous cell
    result = sentiment_analyzer(headline)[0]
    
    label = result['label']
    confidence = result['score'] * 100
    
    print(f"\nLive News: '{headline}'")
    print(f"AI Prediction: {label.upper()} (Confidence: {confidence:.2f}%)")
```
